```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

servico = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=servico)
```

```python
import pandas as pd

notas_df = pd.read_excel('NotasEmitir.xlsx')
notas_df['CEP'] = notas_df['CEP'].apply(str)

# entrar na página de login

driver.get('file:///home/victor/Área de Trabalho/selenium/18/login.html')

# preencher login e senha

driver.find_element(By.XPATH, '/html/body/div/form/input[1]').send_keys('Victor')
driver.find_element(By.XPATH, '/html/body/div/form/input[1]').get_attribute('value')

driver.find_element(By.XPATH, '/html/body/div/form/input[2]').send_keys('123456')
driver.find_element(By.XPATH, '/html/body/div/form/input[2]').get_attribute('value')

# clicar no botao de fazer login


driver.find_element(By.XPATH, '/html/body/div/form/button').click()

# preencher dados da NF

for i, cliente in enumerate(notas_df['Cliente']):
    filtro = notas_df['Cliente'] == cliente
    # nome
    driver.find_element(By.XPATH, '//*[@id="nome"]').clear()
    driver.find_element(By.XPATH, '//*[@id="nome"]').send_keys(cliente)

    # endereço
    driver.find_element(By.XPATH, '/html/body/div/form/input[2]').clear()
    driver.find_element(By.XPATH, '/html/body/div/form/input[2]').send_keys(notas_df['Endereço'][i])

    # bairro
    driver.find_element(By.XPATH, '//*[@id="formulario"]/input[3]').clear()
    driver.find_element(By.XPATH, '//*[@id="formulario"]/input[3]').send_keys(notas_df['Bairro'][i])

    # município
    driver.find_element(By.XPATH, '//*[@id="formulario"]/input[4]').clear()
    driver.find_element(By.XPATH, '//*[@id="formulario"]/input[4]').send_keys(notas_df['Municipio'][i])

    # cep
    cep = int(notas_df['CEP'][i])
    driver.find_element(By.XPATH, '/html/body/div/form/input[5]').clear()
    driver.find_element(By.XPATH, '/html/body/div/form/input[5]').send_keys(cep)

    # uf
    driver.find_element(By.XPATH, '/html/body/div/form/select').send_keys(notas_df['UF'][i])

    # cpf/cnpj
    driver.find_element(By.XPATH, '/html/body/div/form/input[6]').clear()
    cpf = int(notas_df['CPF/CNPJ'][i])
    driver.find_element(By.XPATH, '/html/body/div/form/input[6]').send_keys(cpf)

    # inscricao
    inscricao = int(notas_df['Inscricao Estadual'][i])
    driver.find_element(By.XPATH, '/html/body/div/form/input[7]').clear()
    driver.find_element(By.XPATH, '/html/body/div/form/input[7]').send_keys(inscricao)

    # descricao
    driver.find_element(By.XPATH, '/html/body/div/form/input[8]').clear()
    driver.find_element(By.XPATH, '/html/body/div/form/input[8]').send_keys(notas_df['Descrição'][i])

    # quantidade
    quantidade = int(notas_df['Quantidade'][i])
    driver.find_element(By.XPATH, '/html/body/div/form/input[9]').clear()
    driver.find_element(By.XPATH, '/html/body/div/form/input[9]').send_keys(quantidade)

    # valor unitario
    valor_un = int(notas_df['Valor Unitario'][i])
    driver.find_element(By.XPATH, '/html/body/div/form/input[10]').clear()
    driver.find_element(By.XPATH, '/html/body/div/form/input[10]').send_keys(valor_un)
    # valor total
    valor_total = int(notas_df['Valor Total'][i])
    driver.find_element(By.XPATH, '/html/body/div/form/input[11]').clear()
    driver.find_element(By.XPATH, '/html/body/div/form/input[11]').send_keys(valor_total)

    # clicar em emitir nota fiscal
    driver.find_element(By.XPATH, '/html/body/div/form/button').click()
```
